# Tasks

### 1) Find GDP Growth of India from 1960 - 2016
### 2) Create a new columns named as GDP which shows the GDP growth of each country for that year
### 3) Create a graph of GDP Growth of India from 1960 - 2016

```python
df1 = df.copy()
```

```python
df1 = df1[df1['Country Name']=='India']
df1.head()
```

```python
df2 = df1.values
df2
```

```python
gdp = [0]
for i in range(1,len(df2)):

  prev = df2[i-1][3]
  current = df2[i][3]
  gdp.append(round(((current - prev)/prev)*100,2))

gdp
```

```python
df2 = df1.assign(GDP_Growth = gdp)
```

```python
df2.head()
```

```python
df3 = df.copy()
```

```python
final = []
for country in df3['Country Name'].unique():
  df4 = df3[df3['Country Name']==country]
  data1 = df4.values
  gdp1 = [0]

  for i in range(1,len(data1)):
    prev1 = data1[i-1][3]
    current1 = data1[i][3]
    gdp1.append(round(((current1 - prev1)/prev1)*100,2))
  df4 = df4.assign(GDP_Growth = gdp1)
  final.append(df4)
```

```python
df3 = pd.concat(final, axis=0)
```

```python
df3
```

```python
df_India = df3[df3['Country Name']=='India']
fig1 = px.line(df_India, x = 'Year', y = 'GDP_Growth', title = 'India GDP Growth', color = 'Country Name')
fig1
```

```python
c1 = df3[df3['Country Name'] == 'China']
c2 = df3[df3['Country Name'] == 'India']
df_pr = pd.concat([c1,c2],axis = 0)
fig = px.line(df_pr, x = 'Year', y = 'Value', title = 'GDP Comparison | India & China', color = 'Country Name')
fig
```
